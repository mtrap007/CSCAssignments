Create a basic line plot of the function y = sin(x) over the interval [0, 2*pi]. Label the axes and add a title.


```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.figure(figsize=(8, 5))  # Optional: set figure size
plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Plot of y = sin(x)'
          )

# Show the plot
plt.grid(True)  # Optional: add grid
plt.show()
```
